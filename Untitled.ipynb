# Metabolite Dysregulation in COVID-19 Patients

## Overview

Small molecules produced by cellular metabolism, called metabolites, are essential for life. They play fundamental roles, including energy production, serving as building blocks for proteins, fats, and nucleic acids, and contributing to cellular structure.

Because SARS-CoV-2 (the virus responsible for COVID-19 disease) needs the cell's machinery to replicate inside host cells, it diverts some of the cell's metabolic processes for its own use. This can lead to a condition called metabolite dysregulation.

In this notebook, we will analyze samples from 158 patients with and without COVID-19 to identify metabolites dysregulated by SARS-CoV-2 infection in humans. The data were obtained from 158 patients at a local hospital in Zacatecas, Mexico, who were classified by PCR tests as COVID-19 positive (123 subjects) or COVID-19 negative (35 subjects). Furthermore, positive patients were classified by their level of severity as G2 (not-hospitalized), G3 (hospitalized), or G4 (intubated).

## Metabolic Differences in COVID-19 vs. Healthy Controls

### t-test (two independent samples)

```python
ttest_results = []

for metabolite in metabolites:
    ttest_result = pg.ttest(
        df_metabolites[metabolite][data.Positivo==0],
        df_metabolites[metabolite][data.Positivo==1], 
        alternative='two-sided',
        confidence=0.99
    )
    ttest_results.append(ttest_result)

met_t_significant = []
ttest_results_significant = []

for metabolite, ttest_result in zip(metabolites, ttest_results):
    if ttest_result['p-val'].values[0] < .01 and ttest_result['cohen-d'].values[0]>.5:
        met_t_significant.append(metabolite)
        ttest_results_significant.append(ttest_result)

met_t_significant.sort(
    key=lambda met: df_metabolites[met][data.Positivo==1].mean() - df_metabolites[met][data.Positivo==0].mean()
)
```

```python
df_met_scaled = MinMaxScaler().fit_transform(df_metabolites)
df_met_scaled = pd.DataFrame(df_met_scaled, columns=df_metabolites.columns)
fig, ax = plt.subplots(figsize=(8, 12))

plt.setp([fig, ax], facecolor='.99')

for i, met in enumerate(met_t_significant):
    ax.plot([df_met_scaled[met][data.Positivo==0].mean(), df_met_scaled[met][data.Positivo==1].mean()], 
            [i,i], color='black', lw=1.5, zorder=5)
    ax.scatter([df_met_scaled[met][data.Positivo==0].mean(), df_met_scaled[met][data.Positivo==1].mean()], 
               [i,i], c=['#0C6291', '#D62839'], marker='o', linewidths=2, s=60, zorder=5)

ax.tick_params(left=False, bottom=False)
ax.set_yticks(range(len(met_t_significant)), met_t_significant)
ax.grid(axis='x', color='#000004', lw=2, alpha=0.15)
ax.grid(axis='y', color='#000004', lw=1, alpha=0.05)
ax.set(xlim=(0, 1), ylim=(-1, len(met_t_significant)))
ax.spines['bottom'].set_visible(False)
ax.xaxis.set_major_formatter(plt.FuncFormatter(label_minmax))
for label in ax.get_xticklabels():
    if label.get_text() in ['Min', 'Max']:
        label.set_fontsize(14)
plt.setp(ax.get_xticklabels() + ax.get_yticklabels(), color='#052639')

l1, = plt.plot([], [], ls='', marker='o', color='#0C6291', label='Negative', 
               markersize=8)
l2, = plt.plot([], [], ls='', marker='o', color='#D62839', label='Positive', 
               markersize=8)
ax.legend(loc='upper right', ncol=2, 
          fancybox=True, borderpad=1, title='Means')

fig.suptitle('Metabolites dysregulated ($p<0.01$) *', 
             ha='left', x=-0.1, y=0.92, fontweight='bold')

fig.text(s='* All metabolites were scaled from 0 to 1', x=0.95, y=0.05, ha='right')

plt.savefig('./images/metabolites_disregulated.png', bbox_inches='tight')
```

### Effect-size (Cohen's d)

```python
ttests_sign = pd.concat(ttest_results_significant)
ttests_sign.index = met_t_significant
ttests_sign = ttests_sign[['T', 'p-val', 'CI99%' ,'cohen-d']].sort_values(by='p-val')

ttests_sign.sort_values(by='T').style.background_gradient(subset=['cohen-d'], cmap='Blues', vmin=0.35, vmax=1.5)
```

```python
fig, axs = plt.subplots(len(met_t_significant), 1, figsize=(10, 1.35*len(met_t_significant)),  
                        gridspec_kw={'hspace': 0.3, 'top':0.9}, sharex=False)
plt.setp(axs, facecolor='none')

for ax, met in zip(axs.flat, met_t_significant):
    sns.kdeplot(df_metabolites[met][data.Positivo==0], fill=True, 
                color='#E35226', ax=ax, label='Negative')
    sns.kdeplot(df_metabolites[met][data.Positivo==1], fill=True, 
                color='#0093B8', ax=ax, label='Positive')
    ax.spines[['top', 'left', 'right']].set_visible(False)
    ax.spines['bottom'].set(color='#052639', lw=2, alpha=0.1)
    ax.tick_params(axis='x', labelcolor='#052639', color='#052639', labelsize=9)
    ax.set(ylabel='', yticks=[], xlabel='')

    drange = df_metabolites[met].max() - df_metabolites[met].min()
    ax.set_xlim(df_metabolites[met].min() - 0.1*drange, 
                df_metabolites[met].max() - 0.1*drange)
    ax.set_xticks([df_metabolites[met].min(), 
                   df_metabolites[met].max() - 0.2*drange])
    ax.annotate(met, xy=(0, 0.15), ha='center', va='center', 
                xycoords='axes fraction', color=('#052639', 0.35), fontsize=11, 
                fontweight='regular')
    d = ttests_sign.loc[met, 'cohen-d']

    ax.annotate(f"d={d:.2f}", xy=(0.95, 0.45), xycoords='axes fraction', 
                ha='center', va='center', color='#031925', fontweight='bold',
                bbox={'boxstyle':'round,pad=0.7','facecolor':'#F0F7F9'})

axs[0].annotate("Cohen's d", xy=(0.95, 1), xycoords='axes fraction', 
                ha='center', color='#07324B', fontweight='bold', fontsize=11)
ax_legend = axs[-1].inset_axes([0, -1, 1, 0.8])
ax_legend.axis('off')
ax_legend.legend(handles=axs[0].get_children()[:2], 
                 labels=['Positive', 'Negative'], ncol=2, 
                 loc='lower center', fancybox=True)

fig.suptitle('Changes in Metabolite Levels Between COVID-19 and Healthy Patients', x=0.1, y=0.91, 
             ha='left')

plt.savefig('./images/covid_vs_healthy_metabolite_levels.png', bbox_inches='tight')
```

### Results

#### Downregulated metabolites

```python
ttests_sign[ttests_sign['T'] < 0]
```

Fourteen metabolites showed significantly decreased levels in COVID-19 patients compared to healthy controls ($p < 0.01$), acording with the t-test for two independent samples. The effect sizes were moderate to large, ranging from 0.52 to 1.02 according to Cohen's d.

Overall, downregulation of these metabolites in COVID-19 patients suggests potential disruptions in various cellular processes, energy metabolism, and possibly the immune response. However, more research is required to fully understand the specific implications for disease progression and potential therapeutic interventions.

Overall, upregulation of these metabolites in COVID-19 patients suggests potential increases in cellular activity, alterations in energy production, and possibly inflammatory responses.
